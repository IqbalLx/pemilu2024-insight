```python
partai_master_df = pd.read_csv("./data/partai.csv")
```

```python
partai_master_df = partai_master_df[['id_pilihan', 'nama']]
partai_master_df = partai_master_df.rename(columns={'id_pilihan': 'partai_id'})
```

```python
partai_master_df
```

```python
partai_df_agg = partai_df.groupby('partai_id')['jumlah_suara_total'].sum().reset_index()
partai_df_agg = partai_df_agg.sort_values(by='jumlah_suara_total', ascending=False)
```

```python
sum_all_votes = partai_df['jumlah_suara_total'].sum()
partai_df_agg['jumlah_suara_perc'] = partai_df_agg.apply(lambda row: (row['jumlah_suara_total'] / sum_all_votes) * 100, axis=1)
```

```python
partai_df_agg_joined = partai_df_agg \
    .set_index('partai_id') \
    .join(
        partai_master_df.set_index('partai_id'), 
        on='partai_id', how='inner'
    )\
    .sort_values(by='jumlah_suara_total', ascending=True)
```

```python
partai_df_agg_joined
```

```python
plt.figure(figsize=(10, 8))
bars = plt.barh(partai_df_agg_joined['nama'], partai_df_agg_joined['jumlah_suara_perc'], color='skyblue')
plt.xlim(0, 20)  # Setting the x-axis limit from 0 to 20
plt.xlabel('Percentage of Total Votes')
plt.ylabel('Party Name')
plt.title('Percentage of Total Votes by Party')
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Adding gridlines for better readability

# Adding vertical line at x=4
plt.axvline(x=4, color='red', linestyle='--', label='Parliamentary Threshold (4%)')

# Annotating each bar end with the rounded percentage value
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, '{:.2f}%'.format(width), ha='left', va='center')

plt.tight_layout()  # Adjust layout to prevent cropping labels
plt.show()
```
